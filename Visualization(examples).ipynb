```python
import matplotlib.pyplot as plt
import numpy as np
import os
```

```python
def make_group(queue):
    packet = []
    tmp = []
    v = queue.pop(0)
    tmp.append(v)

    while(len(queue)>0):
        vv = queue.pop(0)
        if v+1 == vv:
            tmp.append(vv)
            v = vv
        else:
            packet.append(tmp)
            tmp = []
            tmp.append(vv)
            v = vv

    packet.append(tmp)
    return packet

def drawing(data, label, column) :
    print(column)
    plt.figure(figsize=(30,5))
    plt.plot(data[:,column], markersize=1)
    
    #---# interval of real anomaly value #---#
    for lb in label:
        a = np.linspace(lb[0], lb[-1])
        plt.fill_between(a, min(data[:,column]), max(data[:,column]), color='green', alpha=0.3)

    plt.show()
    plt.clf()
```

```python
#---# visualization #---#
data_folder = '/content/TimeSeries-Anomaly-Detection-Dataset/data'
dataset = 'SMAP' # MSL, SMAP
dir_location = os.path.join(data_folder, dataset)

label = 'A-4_labels.npy'
A_4_label = np.load(os.path.join(dir_location, label))
test = 'A-4_test.npy'
A_4_test = np.load(os.path.join(dir_location, test))

# print(len(np.where(A_4_label[:,0]==1)[0]))
anomaly_point = np.where(A_4_label[:,0]==1)[0]
anomaly_grouping = make_group(anomaly_point.tolist())

for col in range(len(A_4_test[0,:])):
  drawing(A_4_test, anomaly_grouping, col)
```
